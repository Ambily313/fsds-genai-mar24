# Object Approach in Seaborn

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = [1,2,3,4,5]
y = [10,5,2,6,3]
```

```python
plt.scatter(x,y)

plt.show()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,2,2]) #rect : tuple (left, bottom, width, height)

ax.plot(x,y)
ax.set_xlabel('X values')
ax.set_title('X vs Y')

```

```python
fig,ax = plt.subplots(3, 2)


ax[0,0].scatter(x,y)


ax[0,1].bar(x,y)

ax[1,0].plot(x,y)

```
